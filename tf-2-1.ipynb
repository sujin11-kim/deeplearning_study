```python
import numpy as np
import tensorflow as tf
```

```python
#학습 data set
x_train = [1, 2, 3, 4]
y_train = [0, -1, -2, -3]
```

```python
tf.model = tf.keras.Sequential() #tf 모델 정의
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=1)) 
# units : output shape, input_dim : input shape 
```

```python
sgd = tf.keras.optimizers.SGD(lr=0.1)  
# SGD : standard gradient descendent 알고리즘 이용, lr : 학습 비율
tf.model.compile(loss='mse', optimizer=sgd)  
# mse : mean_squared_error, 1/m * sig (y'-y)^2
```

```python
tf.model.summary() #모델 확인
```

```python
tf.model.fit(x_train, y_train, epochs=200) #data set으로 200번 학습
```

```python
y_predict = tf.model.predict(np.array([5, 4])) #input : 5,4를 넣어 예측값을 얻는다
print(y_predict)
```
